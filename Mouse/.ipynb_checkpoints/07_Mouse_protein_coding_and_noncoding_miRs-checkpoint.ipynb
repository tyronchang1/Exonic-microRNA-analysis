## Plot (protein-coding)

```python
hg_intramiR_coding_df=human_intramiR_gene_type.loc[human_intramiR_gene_type['gene_type']=="protein-coding gene"]
```

```python
hg_intramiR_coding=hg_intramiR_coding_df.groupby(hg_intramiR_coding_df['host_gene_region']).size().reset_index(name='Count')
```

```python
hg_intramiR_coding
```

```python
hg_exonic_intramiR_coding= hg_intramiR_coding[hg_intramiR_coding['host_gene_region']=='exons'].sum()
hg_intronic_intramiR_coding= hg_intramiR_coding[hg_intramiR_coding['host_gene_region']=='introns'].sum()
```

```python
#Human miR loci distribution pie Chart
labels = ['Exons miRNAs', 'Intronic miRNAs']
sizes = [hg_exonic_intramiR_coding['Count'], hg_intronic_intramiR_coding['Count']]
colors = ['skyblue','lightcoral', ]
plt.figure(figsize=(15, 8)) 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Type of human intragenic coding (intronic vs exonic miRNAs) miRNome')
plt.savefig(f'{current_dir}/Pie_charts/Type of human intragenic coding (intronic vs exonic miRNAs) miRNome.eps')
plt.show()
```

```python
##save the tables for MS
hg_intramiR_coding_df.to_excel(f'{current_dir}/MS_tables/Human_Table_protein_coding_Fig2C.xlsx', index=False)
```

## Plot (non-coding)

```python
hg_intramiR_noncoding_df=human_intramiR_gene_type.loc[human_intramiR_gene_type['gene_type']=="non-coding gene"]
```

```python
hg_intramiR_noncoding=hg_intramiR_noncoding_df.groupby(hg_intramiR_noncoding_df['host_gene_region']).size().reset_index(name='Count')
```

```python
hg_intramiR_noncoding
```

```python
hg_exonic_intramiR_noncoding= hg_intramiR_noncoding[hg_intramiR_noncoding['host_gene_region']=='exons'].sum()
hg_intronic_intramiR_noncoding= hg_intramiR_noncoding[hg_intramiR_noncoding['host_gene_region']=='introns'].sum()
```

```python
#Human miR loci distribution pie Chart
labels = ['Exons miRNAs', 'Intronic miRNAs']
sizes = [hg_exonic_intramiR_noncoding['Count'], hg_intronic_intramiR_noncoding['Count']]
colors = ['skyblue','lightcoral', ]
plt.figure(figsize=(15, 8)) 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Type of human intragenic noncoding (intronic vs exonic miRNAs) miRNome')
plt.savefig(f'{current_dir}/Pie_charts/Type of human intragenic noncoding (intronic vs exonic miRNAs) miRNome.eps')
plt.show()
```
